```python
s = 'sadsjkldagvknhdjch adllcagg'

def most_frequent(s: str):

    l = {l:0 for l in set(s)}

    for c in s:
        l[c] += 1

    l = sorted(l.items(), key=lambda r: r[1])
    l.reverse()

    l = list([i[0] for i in l])
    print(l)
    return(l)


most_frequent(s)
```
