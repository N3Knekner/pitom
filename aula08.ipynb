```python
import pandas as pd
import json
import sqlite3 as sql
```

```python
conexao = sql.connect('globalstore_normalizada.db')
```

```python
markets_ = pd.read_sql(con=conexao, sql="SELECT * from customer join city using(cty_sequence) join region using(reg_sequence) where market in ('LATAN', 'US', 'Canada')")
markets_
```

```python
orders_ = pd.read_sql(con=conexao, sql=f"SELECT sub_sequence from items join orders using(ord_sequence) join customer using(cst_sequence) join city using(cty_sequence) join region using(reg_sequence) limit 2")
orders_
```

```python
category_ = pd.read_sql(con=conexao, sql=f"""
SELECT distinct orders.* from orders 
join customer using(cst_sequence) 
join items using(ord_sequence) 
join products using (pro_sequence) 
join subcategory using(sub_sequence) 
join category using (cat_sequence) 
join city using(cty_sequence) 
join region using(reg_sequence) 
where market like 'EMEA' 
--and strftime('%y',order_date) = 2014
and category = 'Furniture'
""")
category_
```
