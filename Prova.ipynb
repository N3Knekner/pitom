### Questão 1 - Fermat

```python
def check_fermat(a: float, b: float, c: float, n: int):
    if n <= 2:
        return
    
    if a**n + b**n == c**n:
        print('Fermat estava errado!')
    else:
        print('Ok, ele estava certo.')

check_fermat(1,2,3,4)

# Alternativa funcional
def check_fermat(a: float, b: float, c: float, n: int):
    if n <= 2:
        return None

    return a**n + b**n == c**n


if r := check_fermat(1,2,3,1):
    print('Fermat estava errado!')
elif r == None:
    print('Regra não se aplica a exponentes <= a 2')
else:
    print('Ok, ele estava certo.')
```

### Questão 2 - Manipulação com listas

```python
# cria uma lista contendo TODOS os numeros de 104 até 10
lista = list(range(10, 105))
lista.sort(reverse=True)
# exibe a lista
print(lista)

# retorna a soma de todos os numeros em POSIÇOES pares
def soma_numeros_pares(l: [int]) -> int:
    return sum([n for i, n in enumerate(l) if i % 2 == 0])

print(soma_numeros_pares(lista))

# Como está na mesma frase, então fica entendido que a função deve retornar os 5 ultimos números também:
def soma_numeros_pares(l: [int]) -> int:
    return sum([n for i, n in enumerate(l) if i % 2 == 0]), l[-5:]

print(soma_numeros_pares(lista))
```

### Questão 5 - Número de telefone

```python
# Exemplo não bate com o enunciado

# implementação de celular:

def corrige_celular(s: str):
    s = s.replace('-', '')
    if len(s) == 8:
        s = '9'+s
    
    return s

print('Celular corrigido: ')
print(corrige_celular('98765-4488'))
print(corrige_celular('8765-4488'))
print(corrige_celular('987654488'))
print(corrige_celular('87654488'))


# implementação de telefone:

def corrige_telefone(s: str):
    s = s.replace('-', '')
    if len(s) == 7:
        s = '3'+s
    
    return s

print('Telefone corrigido: ')
print(corrige_telefone('3942-1202'))
print(corrige_telefone('942-1202'))
print(corrige_telefone('39421202'))
print(corrige_telefone('9421202'))

# Formata o numero
def formata(s: str):
    return s[:-4] + '-'+ s[-4:]

print('Formatado: ')
print(formata(corrige_celular('8765-4488')))
print(formata(corrige_telefone('942-1202')))
```

```python
import pandas as pd

vendas = {'registro_produto': ['laptop', 'tablet', 'laptop', 'mesa', 'cadeira', 'mesa', 'tablet', 'cadeira', 'cadeira', 'laptop', 'tablet', 'cadeira'],
        'data': ['03/05/2023', '17/11/2022', '09/08/2022', '14/05/2023', '23/07/2021', '09/09/2021', '19/08/2023', '23/07/2021', '23/07/2021', '09/08/2022', '24/01/2021', '18/03/2023']}

df_vendas = pd.DataFrame(vendas, index=['venda_1', 'venda_2', 'venda_3', 'venda_4', 'venda_5', 'venda_6', 'venda_7', 'venda_8', 'venda_9', 'venda_10', 'venda_11', 'venda_12'])
df_vendas
```

a) Incorpore uma coluna ao df_produtos com a receita total de cada produto por ano, ou seja, uma coluna com o ano 2021, outra com 2022 e outra com 2023 para cada produto.

```python
df_vendas['data'].astype('datetime64[ns]')
df_vendas['year'] = df_vendas['data'].astype('datetime64[ns]').apply(lambda d: d.year)
df_vendas
```

```python
df_vendas_ano = df_vendas.groupby(['registro_produto', 'year'], as_index = False).size()
df_vendas_ano
    
```

```python
anos = df_vendas['year'].drop_duplicates().tolist()
anos = ['preço_'+str(ano) for ano in anos]
anos
```

```python
df_produtos[anos] = 0
df_produtos
```

```python
for row in df_vendas_ano.itertuples():
    df_produtos.loc[df_produtos['registro_produto'] == row.registro_produto, 'preço_'+str(row.year)] = row.size

df_produtos
```

```python
for ano in anos:
    df_produtos[ano] *= df_produtos['preço']

df_produtos
```
